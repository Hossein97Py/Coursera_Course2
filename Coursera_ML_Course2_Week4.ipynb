the reduction of entropy is called information gained

البته ما آنتروپی بعد تقسیم رو به صورت وزن دار حساب میکنیم. یعنی مثلا 10 تا دیتا داریم. با یک تقسیم در نود به دو گروه تقسیم میشن. باید آنتروپی وزن دار هر گروه رو حساب کنیم . این حداقل باشه وزن مون هم میشه کسر داده هایی که در هر شاخه بعد نود هستند.که البته این کاهش آنتروپی میشه میزان آنتروپی اولیه منهای آنتروپی وزن دار شاخه چپ بعلاوه آنتروپی وزن دار شاخه راست.

one-hot encoding:

تا الان فیچرها دو حالت داشتند. حالا اگه یک فیچر سه حالت مختلف داشت بعد تقسیم بندی در یک نود بر اساس اون فیچر ما سه تا ساب برنچ داریم. برای این میتونیم سه فیچر تعریف کنیم که صفر و یک بگیرن. مثلا یک فیچر سه حالت زیاد و متوسط و کم میگیره. میتونیم سه فیچر تعریف کنیم آیا زیاده؟ آیا متوسطه؟ و آیا کمه؟ که اینا فیچرهای باینری هستند چون فقط صفر و یک میگیرن. و از اینا فقط یکی شون یک میشه و بقیه صفر (وان هات)

Continuous valued features:

فیچر وزن رو در نظر بگیرید . میتونیم سگ و گربه ها رو بر اساس وزن تقسیم کنیم. کاری که میکنیم اینه که یک مرز رو در نظر میگیریم و میگیم مثلا دیتایی که وزنش کمتر از مرز هست میشه گربه و اونایی که بیشترن میشن سگ. مهم اینه که چه وزنی رو به عنوان مرز در نظر بگیریم که بیشترین کسب اطلاعات رو داشته باشیم . که میاییم ونقاط میانی بین دیتا ها رو تست میکنیم همشون رو تا بهترین رو پیدا کنیم.

Regression trees:

مثلا با توجه به شکل گوش، شکل صورت و سیبیلو بودن یا نبودن بخواهیم وزن حیوان رو تخمین بزنیم. الگوریتم اینجوریه که با توجه به تقسیم بندی ها وقتی میرسیم به یک برگ میانگین وزن دیتای ترین که به اون برگ رسیده بهش اختصاص داده میشه و دیتای جدید که به اون برگ رسید همون عدد میانگین براش پیش بینی میشه.

فقط اینکه هنگامی که میخواهیم تصمیم بگیریم دیتا رو بر اساس چه فیچری تقسیم کنیم بحث اینفورمیشن گین مطرح نیست بلکه چیزی که برامون مهم میشه اینه که دیتای هم شاخه بعد تقسیم

Using multiple decision trees:

وقتی یک درخت تصمیم داریم خیلی نسبت به تغییرات دیتا حساس میشه. یک راه حل برای روباست کردن سیستم اینه که از چندتا درخت تصمیم استفاده کنیم. بعد نتیجه هر درخت تصمیم رو داریم و بین نتایج بدست آمده از درخت تصمیم های مختلف رای گیری میکنیم. 

Sampling with replacement

اینجوریه که 4 تا میوه مثلا داریم. میندازیم داخل یک کیسه و تصادفی یکی رو برمیداریم. بعد اونو میندازیم تو کیسه و دوباره یکی دیگه برمیداریم. اینجوری وقتی چهاربار این عمل رو تکرار کنیم چون ویت ریپلیسمنت هست ممکنه یک میوه رو چندبار برداریم و یک میوه رو اصلا برنداریم. 

Random forest algorithm:

تعداد 10 تا دیتا داریم. از اینا مثلا 7 تا دیتاست میسازیم بوسیله همون sampling with replacement

بعد با هر دیتاست یک درخت تصمیم میسازیم. در این حالت یک اتفاقی که خیل ممکنه بیفته اینه که فیچری که باهاش تقسیم رو انجام میدیم در گره ریشه و در گره های نزدیک ریشه همشون  یا اکثرشون شبیه هم میشن
